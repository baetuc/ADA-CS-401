# A little bit of NLP never hurts

```python
from nltk.corpus import stopwords

def remove_stop_words(input_text):
    extra_list = ["'s", "n't"]
    output_text = list()

    for tpl in input_text:
        tpl_lower = tpl.lower()
        if (tpl_lower not in extra_list) and (tpl_lower not in (stopwords.words('english'))):
            output_text.append(tpl)

    return output_text
```

```python
punctuation_marks = [".", "?", "!", ":", ";", "-", "(", ")", "[", "]", "...", "\"", ",",'“', '"', "'", "”", '``', "''", '@', '#', "’", '&']

def delete_punctuations(words):
    return list(filter(lambda x: x not in punctuation_marks, words))
```

```python
import re

def preprocess_tweet(tweet):
    to_delete = ['&amp;', 'RT @[a-zA-Z0-9 ]+:', '(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?']
    tweet_processed = tweet
    
    for expression in to_delete:
        tweet_processed = re.sub(expression, '', tweet_processed)
        
    return tweet_processed
```

```python
def extract_and_remove_hashtags(tweet):
    hashtags = [hashtag[:-1] for hashtag in re.findall('#[^\s]+\s', tweet)]
    tweet_removed_hashtags = re.sub('#[^ ]+ ', ' ', tweet)
    
    return tweet_removed_hashtags, hashtags
```

```python
def process_tweet(tweet):
    tweet, hashtags = extract_and_remove_hashtags(tweet)
    
    tweet = preprocess_tweet(tweet)
    words = nltk.word_tokenize(tweet)
    no_stop_words = remove_stop_words(words)
    no_punctuation = delete_punctuations(no_stop_words)
    
    return no_punctuation, hashtags
```

```python
def count_words_and_hashtags(dataset):
    count_words_dict = {}
    count_hashtags_dict = {}
    
    for tweet in dataset.text:
        words, hashtags = process_tweet(tweet)
        words = list(map(str.lower, words))
        hashtags = list(map(str.lower, hashtags))
        
        
        for word in words:
            if word in count_words_dict:
                count_words_dict[word] += 1
            else: 
                count_words_dict[word] = 1
                
        for hashtag in hashtags:
            if hashtag in count_hashtags_dict:
                count_hashtags_dict[hashtag] += 1
            else: 
                count_hashtags_dict[hashtag] = 1
                
    return count_words_dict, count_hashtags_dict
        
```

# Using words emotion valence

```python
EMOTIONS_FILE = os.path.join('.', 'DepecheMood_V1.0', 'DepecheMood_tfidf.txt')

emotions_df = pd.read_csv(EMOTIONS_FILE, sep='\t')
emotions_df[2000:2010]
```

```python
from nltk.corpus import stopwords

def remove_stop_words_pos(input_text):
    extra_list = ["'s", "n't"]
    output_text = []

    for tpl, pos in input_text:
        tpl_lower = tpl.lower()
        if (tpl_lower not in extra_list) and (tpl_lower not in (stopwords.words('english'))):
            output_text.append((tpl, pos))

    return output_text
```

```python
punctuation_marks = [".", "?", "!", ":", ";", "-", "(", ")", "[", "]", "...", "\"", ",",'“', '"', "'", "”", '``', "''", '@', '#', "’", '&']

def delete_punctuations_pos(words):
    return list(filter(lambda x: x[0] not in punctuation_marks, words))
```

```python
def process_tweet_pos(tweet):
    tweet = tweet.lower()
    tweet, hashtags = extract_and_remove_hashtags(tweet)
    
    tweet = preprocess_tweet(tweet)
    words = nltk.word_tokenize(tweet)
    words_pos = nltk.pos_tag(words)
    
    no_stop_words = remove_stop_words_pos(words_pos)
    no_punctuation = delete_punctuations_pos(no_stop_words)
    
    words_unfiltered = [lemmatize_and_join(word, pos) for word, pos in no_punctuation]
    words_final = list(filter(lambda x: x is not None, words_unfiltered))
    
    return words_final
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.corpus import wordnet
lemmatizer = WordNetLemmatizer()
```

```python
def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return ''
```

```python
def get_depeche_mood_pos(treebank_pos):
    if treebank_pos.startswith('J'):
        return 'a'
    return treebank_pos[0].lower()
```

```python
def lemmatize_and_join(word, pos):    
    pos_wordnet = get_wordnet_pos(pos)
    
    if pos_wordnet == '':
        return None # not a part of speech
    
    lemmatized = lemmatizer.lemmatize(word, pos_wordnet)
    pos_depeche_mood = get_depeche_mood_pos(pos)
    
    return lemmatized + "#" + pos_depeche_mood
```

```python
def initialize_scores():
    score_dict = {}
    
    for emotion in emotions_df.columns[1:]:
        score_dict[emotion] = 0
        
    return score_dict
```

```python
def update_scores(word_pos, score_dict):
    word_emotions_scores = dict(emotions_df[emotions_df['Lemma#PoS'] == word_pos].iloc[0])
    del word_emotions_scores['Lemma#PoS']
    
    for emotion, score in word_emotions_scores.items():
        score_dict[emotion] += score
```

```python
def get_tweet_scores(tweet):
    score_dict = initialize_scores()
    number_words_found = 0
    emotions_words = set(emotions_df['Lemma#PoS'])
    
    words_filtered_pos = process_tweet_pos(tweet)
    
    for word_pos in words_filtered_pos:
        if word_pos not in emotions_words:
            continue
            
        print(word_pos)
            
        update_scores(word_pos, score_dict)
        number_words_found += 1
        
    if number_words_found == 0:
        score_dict_normalized = None
    else:
        score_dict_normalized = {key:value/number_words_found for key, value in score_dict.items()}
    
    return score_dict, score_dict_normalized
```

```python
condensed_final = condensed_dataset.copy()
emotions = set(emotions_df.columns) - {'Lemma#PoS'}

new_columns_dict = {}
for emotion in emotions:
    new_columns_dict[emotion] = []
    new_columns_dict[emotion + '_avg'] = []

    
for tweet in condensed_final.text:
    scores, normalized_scores = get_tweet_scores(tweet)
    
    for emotion, score in scores.items():
        new_columns_dict[emotion].append(score)
        
    if normalized_scores is None:
        continue
    
    for emotion, normalized_score in normalized_scores.items():
        new_columns_dict[emotion + '_avg'].append(normalized_score)
        
for new_column_name, new_column_values in new_columns_dict.items():
    series = pd.Series(new_column_values, index=condensed_final.index)
    condensed_final[new_column_name] = series
```
